```python
#Scrapping total assets for each company on yahoo finance using the method proposed by IzzyAnalytics 
Total_Ass = {}
Total_Lib = {}
PB_no = 0

for i in updated_stock_screen.index:
    try:
        balance_sheet = 'https://finance.yahoo.com/quote/{}/financials?p={}' #Accessing the companies financials
        stock=str(updated_stock_screen["Ticker"][i]) #Using the ticker column data to search each company
        response = requests.get(balance_sheet.format(stock, i),headers={'User-Agent': 'Custom'}) #requesting the balance sheet
        soup = BeautifulSoup(response.text, 'html.parser') #Calling BS4
        pattern = re.compile(r'\s--\sData\s--\s')
        script_data = soup.find('script', text = pattern).contents[0]
        start = script_data.find("context")-2
        json_data = json.loads(script_data[start:-12])
        annual = json_data['context']['dispatcher']['stores']['QuoteSummaryStore']['balanceSheetHistory']['balanceSheetStatements']
        TA = annual[0]['totalAssets']['raw'] #Getting the latest raw data needed 
        PB_no+=1
        Total_Ass[PB_no] = [TA] #Saving data to a new column
        print(TA)
    except KeyError:
        print(0)
        pass
    except ValueError:
        print(0)
        pass
    except IndexError:
        print(0)
        pass
    except TypeError:
        print(0)
        pass
        Total_Ass_df = pd.DataFrame.from_dict(Total_Ass, orient = 'index', columns = ['TotalAssets']) #Saving data to a new column
        example = pd.concat([updated_stock_screen, Total_Ass_df], axis=1, sort=False).reset_index(drop=True) #Saving data to a new column
        example['TotalAssets'] = example['TotalAssets'].shift(periods=-1)
        example.drop(example.tail(1).index,inplace=True)
```

```python
for i in example.index:
    try:
        stock=str(example["Ticker"][i])
        response = requests.get(balance_sheet.format(stock, i),headers={'User-Agent': 'Custom'})
        soup = BeautifulSoup(response.text, 'html.parser')
        pattern = re.compile(r'\s--\sData\s--\s')
        script_data = soup.find('script', text = pattern).contents[0]
        start = script_data.find("context")-2
        json_data = json.loads(script_data[start:-12])
        annual = json_data['context']['dispatcher']['stores']['QuoteSummaryStore']['balanceSheetHistory']['balanceSheetStatements']
        TD = annual[0]['totalLiab']['raw']
        PB_no+=1
        Total_Lib[PB_no] = [TD]
        print(TD)
    except KeyError:
        print(0)
        pass
    except ValueError:
        print(0)
        pass
    except IndexError:
        print(0)
        pass
    except TypeError:
        print(0)
        pass
    Total_Lib_df = pd.DataFrame.from_dict(Total_Lib, orient = 'index', columns = ['TotalLiab']).reset_index(drop=True)
    result2 = example.join(Total_Lib_df, how="outer")
    result2['TotalAssets'] = result2['TotalAssets'].shift(periods=-1)
    result2 = result2.loc[(result2 != 0).all(axis=1), :]
    result2
```

```python
FINAL = pd.read_excel(r"C:\Users\CHARL\OneDrive\Documents\City Uni Data Science modules\Disso\Final disso\Final stock screener df (yahoo data).xlsx", sheet_name='Sheet1')
FINAL.head()

#Performing the value investing appraisal techniques

#P/B Ratio Cal
for ind, row in FINAL.iterrows():
    FINAL.loc[ind, "BVPS"] = row['TotalAssets'] - row['TotalLiab'] 
for ind, row in FINAL.iterrows():
    FINAL.loc[ind, "BVPS1"] = row["BVPS"] / row['SharesOustanding']
for ind, row in FINAL.iterrows():
    FINAL.loc[ind, "P/B Ratio"] = row['SharePrice'] / row['BVPS1'] /100
#Current Ratio Cal
for ind, row in FINAL.iterrows():
    FINAL.loc[ind, "Current Ratio"] = row['CurrentAssets'] / row['CurrentLiab'] 
#D/E Ratio Cal
for ind, row in FINAL.iterrows():
    FINAL.loc[ind, "Total Debt"] = row['LTDebt'] + row['STDebt']
for ind, row in FINAL.iterrows():
    FINAL.loc[ind, "D/E Ratio"] = row["Total Debt"] / row['TotalAssets']
#P/S Ratio Cal
for ind, row in FINAL.iterrows():
    FINAL.loc[ind, "P/S Ratio"] = row["Market Cap"] / row["YearlyRev (2020)"]
#P/E ratio Cal
for ind, row in FINAL.iterrows():
    FINAL.loc[ind, "EPS"] = row["YearlyNet (2020)"] / row['SharesOustanding']
for ind, row in FINAL.iterrows():
    FINAL.loc[ind, "P/E Ratio"] = row["SharePrice"] / row[ "EPS"]
#Gross profit margin % cal 2020
for ind, row in FINAL.iterrows():
    FINAL.loc[ind, "Gross Profit 2020"] = row["YearlyRev (2020)"] - row["YearlyCOCG (2020)"]
for ind, row in FINAL.iterrows():
    FINAL.loc[ind, "GPM % 2020"] = row["YearlyCOCG (2020)"] / row["YearlyRev (2020)"] * 100
#Gross profit margin % cal 2020
for ind, row in FINAL.iterrows():
    FINAL.loc[ind, "Gross Profit 2019"] = row["YearlyRev (2019)"] - row["YearlyCOCG (2019)"]
for ind, row in FINAL.iterrows():
    FINAL.loc[ind, "GPM % 2019"] = row["YearlyCOCG (2019)"] / row["YearlyRev (2019)"] * 100
#Gross profit margin % cal 2018
for ind, row in FINAL.iterrows():
    FINAL.loc[ind, "Gross Profit 2018"] = row["YearlyRev (2018)"] - row["YearlyCOCG (2018)"]
for ind, row in FINAL.iterrows():
    FINAL.loc[ind, "GPM % 2018"] = row["YearlyCOCG (2018)"] / row["YearlyRev (2018)"] * 100
    FINAL['GPM % 2018'].iloc[1] = 0
#calculating the average GPM% OF the last three financial calander years    
for ind, row in FINAL.iterrows():
    FINAL.loc[ind, "Average GPM % 2018-2020"] = (row["GPM % 2018"] + row["GPM % 2019"] + row["GPM % 2020"]) / 3
#checking if the latest net income for recent calander year and if it was positive or negative since a differnce calculation will be needed.    
FINAL['Latest Net Income Pos or Neg?'] = np.where(FINAL['YearlyNet (2020)'] >= 0,  "Positive", "Negative")
FINAL['Average Net Income Rise/Fall % 2018-2020'] = np.where(FINAL['Latest Net Income Pos or Neg?'] == True, ((FINAL['YearlyNet (2020)'] - FINAL['YearlyNet (2019)']) / ((FINAL['YearlyNet (2019)'])) + (FINAL['YearlyNet (2019)'] - FINAL['YearlyNet (2018)']) / (FINAL['YearlyNet (2018)'])) / 2 ,((FINAL['YearlyNet (2020)'] - FINAL['YearlyNet (2019)']) / ((FINAL['YearlyNet (2019)'])) + (FINAL['YearlyNet (2019)'] - FINAL['YearlyNet (2018)']) / (FINAL['YearlyNet (2018)'])) / 2)
SL1 = FINAL.drop(columns=['TotalAssets', 'Total Debt', 'TotalLiab',	'EPS','CurrentAssets', 'CurrentLiab',	'LTDebt', 'STDebt',	'YearlyRev (2020)',	'YearlyRev (2019)',	'YearlyRev (2018)',	'YearlyCOCG (2020)', 'YearlyCOCG (2019)',	'YearlyCOCG (2018)', 'YearlyNet (2020)', 'YearlyNet (2019)', 'YearlyNet (2018)', 'BVPS', 'BVPS1', 'Gross Profit 2020', 'GPM % 2020', 'Gross Profit 2019', 'GPM % 2019',	'Gross Profit 2018', 'GPM % 2018'])

#Scoring THE RESULTS OF THE VALUE INVESTING APPRESIAL TECHNIQUES basing targest on the advice of experts

cm = sns.light_palette("green", as_cmap=True)
true = 1
false = 0
SL1['P/B Score'] = np.where(SL1['P/B Ratio'] <= 2, true, false)
SL1['D/E Score'] = np.where(SL1['D/E Ratio'] <= 0.2, true, false)
SL1['Current Ratio Score'] = np.where(SL1['Current Ratio'] >= 1.5, true, false)
SL1['GPM % Score'] = np.where(SL1["Average GPM % 2018-2020"] >= 50, true, false)
SL1['NI Score'] = np.where(SL1["Latest Net Income Pos or Neg?"] >= "Positive", true, false)
SL1['Ownership Score'] = np.where(SL1['InsiderOwnership'] >= 0.30, true, false)
SL1['P/E Score'] = np.where(SL1['P/E Ratio'] <= 10, true, false)
SL1['P/S Score'] = np.where(SL1['P/S Ratio'] <= 10, true, false)
for ind, row in SL1.iterrows():
    SL1.loc[ind, "Total Score"] = row['P/B Score'] + row['D/E Score'] + row['Current Ratio Score'] + row['GPM % Score'] + row['Ownership Score'] + row['P/E Score'] + row['P/S Score'] + row['NI Score'] 
SL2 = SL1.sort_values(by=['Total Score'], ascending=False)
SL3 = SL2[0:5].sort_values(by=['Total Score'],  ascending=False).round(2).reset_index(drop=True)
SL4 = SL3.style.set_properties(**{'background-color': 'black',
                           'color': 'lawngreen',
                          'border-color': 'white'})


SL3.to_excel('C:\\Users\\CHARL\\OneDrive\\Documents\\City Uni Data Science modules\\Disso\Final disso\\Top NLP Screen Results.xlsx')
SL4
```
